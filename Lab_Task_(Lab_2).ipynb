```python
import numpy as np
from skimage import io, color, exposure, img_as_float
import matplotlib.pyplot as plt
```

# Task 1

```python
# Import the "good.jpeg" image
I = io.imread("/content/good.jpeg")
#I = io.imread()


# Normalize the image
I_norm = img_as_float(I)

# Perform arithmetic operations (contrast increase/decrease) on the image
I_add = np.clip(I_norm + 0.5, 0, 1)  # Increase by 0.5
I_sub = np.clip(I_norm - 0.4, 0, 1)  # Decrease by 0.4
I_mul = np.clip(I_norm * 2, 0, 1)    # Multiply by 2
I_div = np.clip(I_norm / 3, 0, 1)    # Divide by 3




```

```python

plt.figure(figsize=(15,15))
plt.subplot(1, 5, 1)
io.imshow(I_norm)
plt.axis("off")
plt.title('Original Image')

plt.subplot(1, 5, 2)
io.imshow(I_add)
plt.axis("off")
plt.title('Increased by 0.5')

plt.subplot(1,5, 3)
io.imshow(I_sub)
plt.axis("off")
plt.title('Deccreased by 0.4')


plt.subplot(1, 5, 4)
io.imshow(I_mul)
plt.axis("off")
plt.title('Multiplication by 2')



plt.subplot(1, 5, 5)
io.imshow(I_div)
plt.axis("off")
plt.title('Division by 3')
```

```python

# Import the "land.jpeg" image
I = io.imread("/content/land.jpeg")

# Normalize the image
I_norm = img_as_float(I)


# perform contrast stretching on the image

hist_org, nbins_org = exposure.histogram(I_norm, nbins=256, normalize = True)

I_stretched =np.clip(4 * (I), 0, 1)

hist_str, nbins_str = exposure.histogram(I_stretched, nbins=256, normalize = True)



plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.plot(nbins_org,hist_org)
plt.title('Original histogram')

plt.subplot(1, 2, 2)
plt.plot(nbins_str,hist_str)
plt.title('Stretched  histogram')



plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
io.imshow(I_norm)
plt.title('Original Image')

plt.subplot(1, 2, 2)
io.imshow(I_stretched)
plt.title('Stretched  Image')




```

```python

# Import the "land.jpeg" image
I = io.imread("/content/land.jpeg")

# Normalize the image
I_norm = img_as_float(I)


I_min, I_max = np.min(I_norm), np.max(I_norm)
new_min, new_max = 0, 1
I_stretched = (I_norm - I_min) * (new_max - new_min) / (I_max - I_min) + new_min


hist_org, nbins_org = exposure.histogram(I_norm, nbins=256, normalize=False)
hist_str, nbins_str = exposure.histogram(I_stretched, nbins=256, normalize=False)

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.plot(nbins_org, hist_org)
plt.title('Original Histogram')

plt.subplot(1, 2, 2)
plt.plot(nbins_str, hist_str)
plt.title('Stretched Histogram')

plt.show()

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
io.imshow(I_norm)
plt.axis("off")
plt.title('Original Image')

plt.subplot(1, 2, 2)
io.imshow(I_stretched)
plt.axis("off")
plt.title('Stretched Image')

plt.show()
```
